```python
import requests
import time
import random

api_key = "RGAPI-82b03950-7989-4530-9485-aa9b3f6f7bd9"
headers = {"X-Riot-Token": api_key}

seen_game_ids = set()
seen_puuids = set()
queue_id = 420  # Ranked Solo/Duo

def get_match(match_id):
    url = f"https://europe.api.riotgames.com/lol/match/v5/matches/{match_id}"
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error getting match {match_id}: {response.status_code}")
        return None

def get_matches_from_puuid(puuid, count=10):
    url = f"https://europe.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids"
    params = {"queue": queue_id, "count": count}
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error getting matches for {puuid}: {response.status_code}")
        return []

# Starte mit existierenden game_ids
with open("initial_game_ids.txt") as f:
    for line in f:
        seen_game_ids.add(line.strip())

game_ids_to_expand = list(seen_game_ids)

while len(seen_game_ids) < 10000 and game_ids_to_expand:
    current_game_id = game_ids_to_expand.pop(0)
    match = get_match(current_game_id)
    time.sleep(1.3)  # 100 calls / 120 sec = 1.2s spacing

    if not match:
        continue

    participants = match.get("metadata", {}).get("participants", [])
    random.shuffle(participants)

    new_puuids = [p for p in participants if p not in seen_puuids][:2]
    seen_puuids.update(new_puuids)

    for puuid in new_puuids:
        match_ids = get_matches_from_puuid(puuid)
        time.sleep(1.3)
        for mid in match_ids:
            if mid not in seen_game_ids:
                seen_game_ids.add(mid)
                game_ids_to_expand.append(mid)

    print(f"Aktuell: {len(seen_game_ids)} Spiele gesammelt")

# Ergebnis speichern
with open("collected_game_ids.txt", "w") as f:
    for gid in seen_game_ids:
        f.write(gid + "\n")

print("Fertig. 100.000 Spiele gespeichert.")
```
